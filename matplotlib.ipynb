```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,20,25,20])

a=['CSE','Mech','Civil','ECE']

plt.pie(y,labels =a)

plt.legend(title= 'For Subjects')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([3,8,1,10])

plt.plot(y,'o:r')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([35,20,25,20])

a=[0.2,0,0,0]

b=['Python','Maths','English','Chemistry']

plt.pie(x,labels=b,explode=a)

plt.show()
```
